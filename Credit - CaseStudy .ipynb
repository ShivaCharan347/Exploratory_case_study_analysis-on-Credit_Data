```python
def data_types(df):
    cat_var =[] # creating an empty list for categorial values
    con_var =[] # creating an empty list for continous values
    
    for col in df:
        if df[col].nunique()<50:
            cat_var.append(col)
        else:
            con_var.append(col)
    return cat_var, con_var 
```

```python
# printing the categorial and continous values in a table
categorical_var, continuous_var = data_types(df)
print(tabulate({"cat_var":categorical_var,"con_var":continuous_var},headers=["Categorical","Continuous"]))
```

### Uni-variate analysis on Categorical Columns

```python
# creating a function for the categorial values and print unique, missing values if any and counting the values
def cat_info(col):
    print(f"Unique values in {col} : {df[col].unique()}")
    print(f"Missing Value counts of {col}: {df[col].isnull().sum()}")
    print(f"Mode of {col}: {df[col].mode()[0]}")
    value_counts=df[col].value_counts()
    print("value counts:", value_counts)
```

```python
# creating a visualization function which helps to visual each categorial values with the help of bar and pie chart
def cat_visual(col):
    value_counts= df[col].value_counts()
    per= df[col].value_counts(normalize=True)*100 #value_counts in percentage
    plt.figure(figsize=(8,6))
    plt.bar(value_counts.index,value_counts.values,width=0.25)
    plt.title(f"value count of{col}")
        
    
    plt.figure(figsize=(8,6))
    plt.pie(per.values,labels=per.index,autopct="%1.1f%%")
    plt.title(f"percentage of {col}")
```

```python
def main_cat(col):
    cat_visual(col)
    cat_info(col)
```

```python
def num_info(col):
    summary=df[col].describe()
    print(summary)
    print(f"median of {col}: {df[col].median()}")
    print(f"mode of {col}: {df[col].mode()[0]}")
    print(f"missing values {col}: {df[col].isnull().sum()}")
    print(f"skewness of {col}:{df[col].skew()}")
    print(f"Kurtosis of {col}: {df[col].kurt()}")
    
def num_visual(col):
    fig,ax=plt.subplots(2,1,figsize=(10,10))
    ax[0].hist(df[col])
    sns.boxplot(y=df[col],ax=ax[1])
    
def main(col):
    num_info(col)
    num_visual(col)
```

```python
main('AMT_INCOME_TOTAL')
```

Outlier Treatment

```python
q1 = df["AMT_INCOME_TOTAL"].quantile(.25)
q3 = df["AMT_INCOME_TOTAL"].quantile(.75)
iqr = q3-q1
lf = q1-1.5*iqr
uf = q3+1.5*iqr
df.loc[df["AMT_INCOME_TOTAL"]>uf,"AMT_INCOME_TOTAL"]=df["AMT_INCOME_TOTAL"].median()
df.loc[df["AMT_INCOME_TOTAL"]<lf,"AMT_INCOME_TOTAL"]=df["AMT_INCOME_TOTAL"].median()
```

```python
main('AMT_INCOME_TOTAL')
```

**Condition 1: Car yes, House Yes and Employes Yes wrt Total Income**
